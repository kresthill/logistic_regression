### Preprocessing for Logistic Regression

```python
from sklearn.impute import SimpleImputer


# Replace zeros with NaN in physiological fields
impossible = ["Glucose", "BloodPressure", "SkinThickness", "Insulin", "BMI"]
data[impossible] = data[impossible].replace(0, np.nan)

# Impute missing values using median
imputer = SimpleImputer(strategy="median")
data[impossible] = pd.DataFrame(
    imputer.fit_transform(data[impossible]),
    columns=impossible
)

# Split features and labels
X = data.drop("Outcome", axis=1)
Y = data["Outcome"].to_numpy()

# Standardize data
scaler = StandardScaler().fit(X)
X_scaled = pd.DataFrame(
    scaler.fit_transform(X),
    columns=X.columns
)

```

### Data Splitting

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(
    X, Y, test_size=0.2, shuffle=True, random_state=42
)

```

### Data Modelling / Evaluation

```python
# define the model
model = LogisticRegression(max_iter=1000)
# train the model
model.fit(X_train, Y_train)
# create the predictions
Y_train_pred = model.predict(X_train)
Y_test_pred = model.predict(X_test)

# ===================================
# Model Evaluation
# ===================================

# build the confusion matrices
confusion_matrix_train = confusion_matrix(Y_train, Y_train_pred)
confusion_matrix_test = confusion_matrix(Y_test, Y_test_pred)

# calculate accuracies
train_accuracy = accuracy_score(Y_train, Y_train_pred)
test_accuracy = accuracy_score(Y_test, Y_test_pred)

# display the results
print('Training Confusion Matrix')
print(confusion_matrix_train)
print()
print(f'Training accuracy = {train_accuracy}')
print()
print('Testing Confusion Matrix')
print(confusion_matrix_test)
print()
print(f'Testing accuracy = {test_accuracy}')
```

```python
# classification report
from sklearn.metrics import classification_report

print(classification_report(Y_test, Y_test_pred))
```

```python
from sklearn.metrics import roc_auc_score

auc = roc_auc_score(Y_test, Y_test_pred)
print("AUC:", auc)
```

```python

C_space = np.linspace(10e-6, 10e-2, 200)
accuracies = []

for C in C_space:
  model = LogisticRegression(penalty = 'l2', C = C, max_iter=1000)
  model.fit(X_train, Y_train)
  Y_test_pred = model.predict(X_test)
  accuracies.append(accuracy_score(Y_test, Y_test_pred))

plt.plot(C_space, accuracies)
plt.xlabel('C')
plt.ylabel('Testing Accuracy')
plt.title('Analysis of Regularisation')

best_model = np.argmax(accuracies)
best_accuracy = accuracies[best_model]
best_C = C_space[best_model]

print(f'Best test accuracy was achieved with C = {best_C}, giving an accuracy of {best_accuracy}.') 
print(f'Without regularisation, we achieved an accuracy of {test_accuracy}.')
print()
```

```python
# define and train model
model = LogisticRegression(penalty = 'l2', C = best_C, max_iter=1000)
model.fit(X_train, Y_train)

def predict_with_threshold(model, X, threshold):
  # calculate the probabilites
  probabilites = model.predict_proba(X)
  # scikit returns the probabilities for each class. However, we only need the second column as there are only two classes
  # and probabilites add up to one. In particular, the second column represents Pr(Y = 1 | X)
  probabilites = probabilites[:, 1]
  # build an array of boolean variables, checking where our probabilities are larger or equal to the threshold
  boolean_threshold = (probabilites >= threshold)
  # since our class labels are 1 and 0, we can simply transform the boolean variables by making them integers
  predictions = boolean_threshold.astype(int)

  return predictions

# define threshold space, initialise lists
thresholds = np.linspace(0, 1, 101)

false_negatives = []
false_positives = []

# calculate number of true positives and negatives
num_of_positives = np.sum(Y_test)
num_of_negatives = len(Y_test) - num_of_positives

for tp in thresholds:
  # predict
  Y_pred = predict_with_threshold(model, X_test, tp)
  #calculate predictions
  cm = confusion_matrix(Y_test, Y_pred)
  # calculate false negative rate and false positive rate
  fnr = cm[1, 0] / num_of_positives
  fpr = cm[0, 1] / num_of_negatives
  # append to lists
  false_negatives.append(fnr)
  false_positives.append(fpr)

# create a plot to see how the rates change with the threshold
fig, ax = plt.subplots(nrows=1, ncols=1)
fig.set_figheight(6)
fig.set_figwidth(10)
fig.suptitle('Investigation of the FNR and FPR against Probability Threshold')

ax.plot(thresholds, false_negatives, 'r', label = 'FNR')
ax.plot(thresholds, false_positives, 'b', label = 'FPR')
ax.set_ylabel('Performance Metrics')
ax.set_xlabel('Probability Thresholds')
ax.legend()
```

```python
# --- choose the best threshold by minimizing total error (FNR + FPR) ---

total_errors = []
for fnr, fpr in zip(false_negatives, false_positives):
    total_errors.append(fnr + fpr)

best_index = np.argmin(total_errors)
best_threshold = thresholds[best_index]
best_total_error = total_errors[best_index]

print("\n--- Best Probability Threshold ---")
print(f"Best threshold = {best_threshold:.3f}")
print(f"Minimum total error (FNR + FPR) = {best_total_error:.4f}")

# Evaluate model at this threshold
Y_best_pred = predict_with_threshold(model, X_test, best_threshold)
cm_best = confusion_matrix(Y_test, Y_best_pred)
acc_best = accuracy_score(Y_test, Y_best_pred)

print("\nConfusion Matrix at Best Threshold:")
print(cm_best)
print(f"Accuracy at best threshold: {acc_best:.4f}")
```

### Ploting the Sigmoid Curve + Best Threshold Marker

```python
from scipy.special import expit   # expit = sigmoid function

# --- Best threshold
best_threshold = 0.520

# Create a range of input scores (logits)
z = np.linspace(-10, 10, 400)

# Compute sigmoid for all points
sigmoid_vals = expit(z)

# Plot sigmoid function
plt.figure(figsize=(10, 6))
plt.plot(z, sigmoid_vals, label="Sigmoid Function Ïƒ(z)")

# Horizontal line at best threshold
plt.axhline(best_threshold, color='red', linestyle='--',
            label=f"Best Threshold = {best_threshold}")

# Find the point on the curve closest to threshold
# (purely visual â€” logistic regression models map logits â†’ probabilities)
threshold_index = np.argmin(np.abs(sigmoid_vals - best_threshold))
z_at_threshold = z[threshold_index]

# Vertical line showing where the sigmoid hits the threshold
plt.axvline(z_at_threshold, color='green', linestyle='--',
            label=f"z â‰ˆ {z_at_threshold:.2f} for p={best_threshold}")

# Labels and title
plt.title("Sigmoid Function with Best Probability Threshold")
plt.xlabel("Logit (z)")
plt.ylabel("Probability Ïƒ(z)")
plt.legend()
plt.grid(True)
plt.show()
```
